```python
object_type_cols = df_18_19_matches.select_dtypes(exclude=['int64', 'float64']).columns
```

```python
df_numeric = df_18_19_matches.drop(columns=object_type_cols)
```

```python
corr_matrix = df_numeric.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation map')
plt.show()
```

```python
#correlation threshold
threshold = 0.8
high_corr_pairs = {}

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            variable1 = corr_matrix.columns[i]
            variable2 = corr_matrix.columns[j]
            correlation_value = corr_matrix.iloc[i, j]
            high_corr_pairs[(variable1, variable2)] = correlation_value

if high_corr_pairs:
    print("Pary zmiennych z wysokim współczynnikiem korelacji (> {})".format(threshold))
    for pair, correlation in high_corr_pairs.items():
        print(f"{pair[0]} i {pair[1]}: {correlation}")
```

```python
#correlation threshold
threshold = 0.8
high_corr_pairs = {}

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if 0.5 < abs(corr_matrix.iloc[i, j]) < 0.8:
            variable1 = corr_matrix.columns[i]
            variable2 = corr_matrix.columns[j]
            correlation_value = corr_matrix.iloc[i, j]
            high_corr_pairs[(variable1, variable2)] = correlation_value

if high_corr_pairs:
    print("Pary zmiennych z umiarkowanym dodatnik współczynnikiem korelacji")
    for pair, correlation in high_corr_pairs.items():
        print(f"{pair[0]} i {pair[1]}: {correlation}")
```
